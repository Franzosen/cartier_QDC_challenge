## Missing Values Strategy
We distinguish three types of missing values:

1. Structural missing → kept (e.g. no repair history)
2. Informative missing → flagged (e.g. AGE)
3. Random missing → treated case by case

We avoid global imputation to preserve signal and prevent bias.

```python
# ==============================
# 9. CHECK DUPLICATES
# ==============================

duplicates = df_agg.duplicated(subset=["CLIENT_ID", "DATE_TARGET"]).sum()
print("Number of duplicates:", duplicates)
```

## Text Sequence Columns

about customer behavior.
They are not processed during the cleaning phase, as they require complex parsing
and transformation.
These columns are preserved and will be used in the feature engineering stage
to extract behavioral patterns (e.g., category progression, price ladder).

```python
# ==============================
# 5. CLEAN SERIAL_NUMBER
# ==============================

placeholder = "4e2e3377c60db4140ae7"

# Replace fake value with NaN
df_trs.loc[df_trs["SERIAL_NUMBER"] == placeholder, "SERIAL_NUMBER"] = np.nan

# Create flag
df_trs["SERIAL_NUMBER_KNOWN"] = df_trs["SERIAL_NUMBER"].notna().astype(int)
```

```python
# ==============================
# 6. WWPRICE FLAG
# ==============================

df_trs["WWPRICE_MISSING"] = (
    (df_trs["ARTICLE_WWPRICE"] == 0) &
    (df_trs["TO_WITHOUTTAX_EUR_CONST"] > 0)
).astype(int)
```
